```python
import pandas as pd
import re
```

```python
Gallery = pd.read_csv("AmericanGallery.csv")
```

```python
Gallery
```

```python
Gallery.rename({"BeginDate":"Begin_Date"}, axis = 1, inplace = True)
Gallery.rename({"EndDate":"End_Date"}, axis = 1, inplace = True)
Gallery.rename({"ActualAge":"Actual_Age"}, axis = 1, inplace = True)
```

```python
Gallery["Nationality"] = Gallery["Nationality"].str.strip("(").str.strip(")")
```

```python
Gallery["Nationality"]
```

```python
Gallery.dtypes
```

```python
Gallery["BeginDate"] = Gallery["BeginDate"].astype(str).str.replace("-", "").astype(float)
```

```python
Gallery["EndDate"] = Gallery["EndDate"].astype(str).str.replace("-", "").astype(float)
```

```python
Gallery.dtypes
```

```python
Gallery["Gender"] = Gallery["Gender"].str.strip("(").str.strip(")")
```

```python
Gallery["Gender"]
```

```python
bad_char = ["(",")",".","c","C","'"," ","S","s"]
```

```python
Gallery['Date'] = Gallery['Date'].str.replace('|'.join(map(re.escape, bad_char)), '')
```

```python
Gallery['Date']
```

```python
Gallery['Date'] = Gallery['Date'].apply(lambda x: x.lstrip('-') if x.startswith('-') else x)
```

```python
# fil = Gallery["Date"][Gallery["Date"].str.contains('-') & (Gallery["Date"].str.len() < 9)]
fil = Gallery["Date"][Gallery['Date'].str.contains('-') & (Gallery['Date'].str.len() < 9)]

```

```python
fil
```

```python
def combine_date_range(date_range):
    start_year, end_year = date_range.split('-')
    
    if len(start_year) == 2:
        start_year = start_year.slice[0:2]
    if len(end_year) == 2:
        end_year = start_year[0:2] + end_year
    
    return f"{start_year}-{end_year}"

fil = fil.apply(combine_date_range)
```

```python
filt = Gallery["Date"][Gallery['Date'].str.contains('-') & (Gallery['Date'].str.len() == 9)]
```

```python
filt
```

```python
df_split = filt.str.split('-', expand=True)

df_split1 = df_split.astype(int)

mean_values = df_split1.mean(axis=1).astype(int)

result = pd.concat([filt, mean_values.rename('Mean')], axis=1)
```

```python
result.drop("Date", axis =1, inplace = True)
```

```python
Gallery["EndDate"] = Gallery["EndDate"].astype("Int64")
```

```python
Gallery["BeginDate"] = Gallery["BeginDate"].astype("Int64")
```

```python
Gallery["Date"] = Gallery["Date"].astype(int)
```

```python
Gallery.dtypes
```

```python
Gallery["ActualAge"] = ""

for index, row in Gallery.iterrows():
    begin_date = row["BeginDate"]
    end_date = row["EndDate"]

    if pd.notna(begin_date) and pd.notna(end_date):
        Gallery.at[index, "ActualAge"] = (end_date - begin_date)
    else:
        Gallery.at[index, "ActualAge"] = "NP"
```

```python
Gallery
```

```python
Gallery["Submission_Age"] = ""

for index, row in Gallery.iterrows():
    begin_date = row["BeginDate"]
    date = row["Date"]

    if pd.notna(begin_date) and pd.notna(date):
        Gallery.at[index, "Submission_Age"] = (date - begin_date)
    else:
        Gallery.at[index, "Submission_Age"] = "NP"
```

```python
Gallery
```

```python
for index, row in Gallery.iterrows():
    if row["Nationality"] == "":
        Gallery.at[index, "Nationality"] = "NP"
```

```python
for index, row in Gallery.iterrows():
    if row["Gender"] == "":
        Gallery.at[index, "Gender"] = "NP"
```
